```python
import pandas as pd

df = pd.read_csv("marketing_campaign.csv",sep=";")

df.info()
df.isnull().sum()

df.drop_duplicates(inplace=True)
```

```python
df.head()
df.info()
df.isnull().sum()
```

```python
df.isnull().sum()
```

```python
df[df['Income'].isnull()]
```

```python
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
df['Age'] = 2024 - df['Year_Birth']
```

```python
df['Age'].describe()
```

```python
df = df[df['Age'] < 100]
```

```python
df['Total_Spending'] = (
    df['MntWines'] +
    df['MntFruits'] +
    df['MntMeatProducts'] +
    df['MntFishProducts'] +
    df['MntSweetProducts'] +
    df['MntGoldProds']
)
```

```python
df['Children'] = df['Kidhome'] + df['Teenhome']
```

```python
df.info()
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Customer Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Education', y='Total_Spending', data=df)
plt.xticks(rotation=45)
plt.title("Spending by Education Level")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Income', y='Total_Spending', data=df)
plt.title("Income vs Total Spending")
plt.show()
```

```python
purchase_channels = df[['NumWebPurchases','NumCatalogPurchases','NumStorePurchases']].sum()

purchase_channels.plot(kind='bar', figsize=(6,4))
plt.title("Purchases by Channel")
plt.ylabel("Total Purchases")
plt.show()
```

```python
sns.countplot(x='Response', data=df)
plt.title("Campaign Response")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Marital_Status', y='Total_Spending', data=df)
plt.xticks(rotation=45)
plt.title("Spending by Marital Status")
plt.show()
```
